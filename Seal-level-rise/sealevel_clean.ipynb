Sea level increases due to increase melting of polar ice caps and the expansion of water due to rise in temperature. Polar ice caps melt due to increase in temperature. Hence, sea level rise is a very good estimate of whether temperature rise/melting of polar ice caps is occuring.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import seaborn as sns
import datetime
import math
```

```python
"""HDR 1 altimeter type 0=dual-frequency  999=single frequency (ie Poseidon-1)
HDR 2 merged file cycle #
HDR 3 year+fraction of year (mid-cycle)
HDR 4 number of observations
HDR 5 number of weighted observations
HDR 6 GMSL (Global Isostatic Adjustment (GIA) not applied) variation (mm) with respect to 20-year TOPEX/Jason collinear mean reference
HDR 7 standard deviation of GMSL (GIA not applied) variation estimate (mm)
HDR* 8 smoothed (60-day Gaussian type filter) GMSL (GIA not applied) variation (mm)  with respect to 20-year mean
HDR* 9 GMSL (Global Isostatic Adjustment (GIA) applied) variation (mm) )  with respect to 20-year mean
HDR 10 standard deviation of GMSL (GIA applied) variation estimate (mm)
HDR* 11 smoothed (60-day Gaussian type filter) GMSL (GIA applied) variation (mm) )  with respect to 20-year mean
HDR* 12 smoothed (60-day Gaussian type filter) GMSL (GIA applied) variation (mm); annual and semi-annual signal removed )  with respect to 20-year mean
"""
data = pd.read_csv('sealeveldata.txt', sep="[ \t]+", header=None)
print(data.head())
data.columns = ["year_fraction","gmsl","sd_gmsl","mean_gmsl","mean_gmsl_gia","sd_mean_gmsl"]
```

```python
data["year"] = [str(x).split('.')[0] for x in data["year_fraction"].values]
data["month"] = ["0."+str(x).split('.')[1] for x in data["year_fraction"].values]
data["month"] = [int((float(x) * 365)/31)+1 for x in data["month"].values]
data["gmsl"].mean()
```

```python
df = pd.DataFrame()

for year in range(1993,2018):
        temp = data[(data["year"]==str(year))].gmsl.mean()
        print(temp)
        df = df.append({'year':year,'gmsl_change':temp},ignore_index=True)
        
df["gmsl_change"] = df["gmsl_change"] - df.loc[0,"gmsl_change"]        
df.to_csv("GMSL_variation_yearwise_1993to2017.csv")
```

```python
df = pd.DataFrame()

for year in range(1993,2018):
    for month in range(1,13):
        temp = data[(data["year"]==str(year)) & (data["month"]==month)].gmsl.mean()
        print(temp)
        df = df.append({'year':year,'month':month,'gmsl_change':temp},ignore_index=True)
```

```python
df.to_csv("GMSL_variation_1993to2017.csv")
```
